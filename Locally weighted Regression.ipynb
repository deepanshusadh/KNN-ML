```python
X=np.mat(X)
```

```python
def getW(q_point,X,tau):
    M=X.shape[0]
    W=np.mat(np.eye(M))
    for i in range(M):
        x=q_point
        W[i][i]=np.exp(((x-X[i])*(x-X[i]).T)/(2*tau*tau))
    return W
```

```python
print(getW(1,X,0.1))
```

```python
def Theta(q_point,X,tau,Y):
    W=getW(q_point,X,tau)
    Y=np.mat(Y)
    theta=np.linalg.pinv(np.dot(W,(np.dot(X.T,X))*(np.dot(W,(np.dot(X.T,Y))))
    return theta
                                
def predict(X,theta):
    return np.dot(X,theta)
```
